```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Show available columns
print("Available columns:")
print(df.columns.tolist())

# Required group columns
group_cols = [
    "AI_Tool_Type",
    "AI_Usage_Level",
    "Risk_Category"
]

# Base aggregation dictionary
agg_dict = {
    "Participants": ("Participant_ID", "count"),
    "Avg_Health_Before": ("Health_Score_Before_AI", "mean"),
    "Avg_Health_After": ("Health_Score_After_AI", "mean"),
    "Avg_Improvement": ("Improvement", "mean"),
    "Median_Improvement": ("Improvement", "median"),
    "Avg_Improvement_Percentage": ("Improvement_Percentage", "mean"),
    "Avg_Satisfaction": ("Satisfaction_Score", "mean")
}

# Add optional columns only if they exist
if "AI_Usage_Hours" in df.columns:
    agg_dict["Avg_AI_Usage_Hours"] = ("AI_Usage_Hours", "mean")

if "AI_Usage_Hours_Per_Day" in df.columns:
    agg_dict["Avg_AI_Usage_Hours"] = ("AI_Usage_Hours_Per_Day", "mean")

if "Exercise_Hours" in df.columns:
    agg_dict["Avg_Exercise_Hours"] = ("Exercise_Hours", "mean")

if "Exercise_Hours_Per_Week" in df.columns:
    agg_dict["Avg_Exercise_Hours"] = ("Exercise_Hours_Per_Week", "mean")

# Create comparison summary
comparison_summary = df.groupby(group_cols).agg(**agg_dict).reset_index()
comparison_summary = comparison_summary.round(2)

print("1. Stronger comparison summary")
display(comparison_summary.head(10))

# Best-performing groups
top_groups = comparison_summary.sort_values(
    by="Avg_Improvement",
    ascending=False
)

print("2. Top 10 best-performing groups")
display(top_groups.head(10))

# Weakest-performing groups
bottom_groups = comparison_summary.sort_values(
    by="Avg_Improvement",
    ascending=True
)

print("3. Bottom 10 weakest-performing groups")
display(bottom_groups.head(10))

# Summary by AI tool type
tool_type_summary = df.groupby("AI_Tool_Type").agg(
    Participants=("Participant_ID", "count"),
    Avg_Improvement=("Improvement", "mean"),
    Avg_Satisfaction=("Satisfaction_Score", "mean"),
    Avg_Improvement_Percentage=("Improvement_Percentage", "mean")
).round(2).sort_values(by="Avg_Improvement", ascending=False)

print("4A. Average improvement by AI tool type")
display(tool_type_summary)

# Summary by AI usage level
usage_agg = {
    "Participants": ("Participant_ID", "count"),
    "Avg_Improvement": ("Improvement", "mean"),
    "Avg_Satisfaction": ("Satisfaction_Score", "mean")
}

if "AI_Usage_Hours" in df.columns:
    usage_agg["Avg_AI_Usage_Hours"] = ("AI_Usage_Hours", "mean")

if "AI_Usage_Hours_Per_Day" in df.columns:
    usage_agg["Avg_AI_Usage_Hours"] = ("AI_Usage_Hours_Per_Day", "mean")

usage_level_summary = df.groupby("AI_Usage_Level").agg(**usage_agg).round(2).sort_values(
    by="Avg_Improvement",
    ascending=False
)

print("4B. Average improvement by AI usage level")
display(usage_level_summary)

# Summary by risk category
risk_category_summary = df.groupby("Risk_Category").agg(
    Participants=("Participant_ID", "count"),
    Avg_Improvement=("Improvement", "mean"),
    Avg_Health_Before=("Health_Score_Before_AI", "mean"),
    Avg_Health_After=("Health_Score_After_AI", "mean")
).round(2).sort_values(by="Avg_Improvement", ascending=False)

print("4C. Average improvement by risk category")
display(risk_category_summary)

# Visuals
tool_avg = df.groupby("AI_Tool_Type")["Improvement"].mean().sort_values()

plt.figure(figsize=(9,5))
tool_avg.plot(kind="barh")
plt.title("Average Health Improvement by AI Tool Type")
plt.xlabel("Average Improvement")
plt.ylabel("AI Tool Type")
plt.show()

usage_avg = df.groupby("AI_Usage_Level")["Improvement"].mean().sort_values()

plt.figure(figsize=(8,5))
usage_avg.plot(kind="barh")
plt.title("Average Health Improvement by AI Usage Level")
plt.xlabel("Average Improvement")
plt.ylabel("AI Usage Level")
plt.show()

risk_avg = df.groupby("Risk_Category")["Improvement"].mean().sort_values()

plt.figure(figsize=(8,5))
risk_avg.plot(kind="barh")
plt.title("Average Health Improvement by Risk Category")
plt.xlabel("Average Improvement")
plt.ylabel("Risk Category")
plt.show()

# Tool type and usage level pivot
tool_usage_pivot = df.pivot_table(
    values="Improvement",
    index="AI_Tool_Type",
    columns="AI_Usage_Level",
    aggfunc="mean"
).round(2)

print("6. Average improvement by AI tool type and usage level")
display(tool_usage_pivot)

tool_usage_pivot.plot(kind="bar", figsize=(10,6))
plt.title("Average Improvement by AI Tool Type and Usage Level")
plt.xlabel("AI Tool Type")
plt.ylabel("Average Improvement")
plt.xticks(rotation=45)
plt.legend(title="AI Usage Level")
plt.show()

# Risk category and usage level pivot
risk_usage_pivot = df.pivot_table(
    values="Improvement",
    index="Risk_Category",
    columns="AI_Usage_Level",
    aggfunc="mean"
).round(2)

print("7. Average improvement by risk category and usage level")
display(risk_usage_pivot)

risk_usage_pivot.plot(kind="bar", figsize=(9,6))
plt.title("Average Improvement by Risk Category and AI Usage Level")
plt.xlabel("Risk Category")
plt.ylabel("Average Improvement")
plt.xticks(rotation=0)
plt.legend(title="AI Usage Level")
plt.show()

# Distribution of improvement
plt.figure(figsize=(9,5))
plt.hist(df["Improvement"], bins=30)
plt.title("Distribution of Health Improvement")
plt.xlabel("Improvement")
plt.ylabel("Number of Participants")
plt.show()

# Health before vs after
plt.figure(figsize=(7,6))
plt.scatter(
    df["Health_Score_Before_AI"],
    df["Health_Score_After_AI"],
    alpha=0.4
)

plt.title("Health Score Before vs After AI")
plt.xlabel("Health Score Before AI")
plt.ylabel("Health Score After AI")
plt.show()

# AI usage hours vs improvement, only if column exists
ai_hours_col = None

if "AI_Usage_Hours" in df.columns:
    ai_hours_col = "AI_Usage_Hours"
elif "AI_Usage_Hours_Per_Day" in df.columns:
    ai_hours_col = "AI_Usage_Hours_Per_Day"

if ai_hours_col:
    plt.figure(figsize=(7,6))
    plt.scatter(
        df[ai_hours_col],
        df["Improvement"],
        alpha=0.4
    )

    plt.title("AI Usage Hours vs Health Improvement")
    plt.xlabel(ai_hours_col)
    plt.ylabel("Improvement")
    plt.show()
else:
    print("Skipped AI usage hours chart because no AI usage hours column was found.")

# Satisfaction vs improvement
plt.figure(figsize=(7,6))
plt.scatter(
    df["Satisfaction_Score"],
    df["Improvement"],
    alpha=0.4
)

plt.title("Satisfaction Score vs Health Improvement")
plt.xlabel("Satisfaction Score")
plt.ylabel("Improvement")
plt.show()

# Power BI-ready grouped dataset
powerbi_group_cols = [
    "Age_Group",
    "Gender",
    "BMI_Category",
    "AI_Tool_Type",
    "AI_Usage_Level",
    "Exercise_Level",
    "Risk_Category",
    "Improvement_Category"
]

powerbi_agg = {
    "Participants": ("Participant_ID", "count"),
    "Avg_Age": ("Age", "mean"),
    "Avg_BMI": ("BMI", "mean"),
    "Avg_Health_Before": ("Health_Score_Before_AI", "mean"),
    "Avg_Health_After": ("Health_Score_After_AI", "mean"),
    "Avg_Improvement": ("Improvement", "mean"),
    "Avg_Improvement_Percentage": ("Improvement_Percentage", "mean"),
    "Avg_Satisfaction": ("Satisfaction_Score", "mean")
}

if "AI_Usage_Hours" in df.columns:
    powerbi_agg["Avg_AI_Usage_Hours"] = ("AI_Usage_Hours", "mean")

if "AI_Usage_Hours_Per_Day" in df.columns:
    powerbi_agg["Avg_AI_Usage_Hours"] = ("AI_Usage_Hours_Per_Day", "mean")

if "Exercise_Hours" in df.columns:
    powerbi_agg["Avg_Exercise_Hours"] = ("Exercise_Hours", "mean")

if "Exercise_Hours_Per_Week" in df.columns:
    powerbi_agg["Avg_Exercise_Hours"] = ("Exercise_Hours_Per_Week", "mean")

if "Improvement_Per_AI_Hour" in df.columns:
    powerbi_agg["Avg_Improvement_Per_AI_Hour"] = ("Improvement_Per_AI_Hour", "mean")

powerbi_grouped_data = df.groupby(powerbi_group_cols).agg(**powerbi_agg).reset_index()
powerbi_grouped_data = powerbi_grouped_data.round(2)

print("12. Power BI grouped data preview")
display(powerbi_grouped_data.head())

# Export files
comparison_summary.to_csv("comparison_summary_ai_health.csv", index=False)
powerbi_grouped_data.to_csv("powerbi_grouped_ai_health_data.csv", index=False)
df.to_csv("cleaned_full_ai_health_data.csv", index=False)

print("Exported files:")
print("comparison_summary_ai_health.csv")
print("powerbi_grouped_ai_health_data.csv")
print("cleaned_full_ai_health_data.csv")

files.download("comparison_summary_ai_health.csv")
files.download("powerbi_grouped_ai_health_data.csv")
files.download("cleaned_full_ai_health_data.csv")
```
